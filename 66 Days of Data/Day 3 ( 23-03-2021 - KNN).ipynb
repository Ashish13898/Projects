- An Easy way to distiguish between classification and regression is to ask wheather there is some kind of continuity in the output. 
    - e.g amount can be 31 or 32 
        english and french are single seperate languages, there is no language that is between english and french
    
- Usually training set and test set should have enough in comman so that our model can predict the values of the new data and be accurate on it

- Accuracy of the model soley depends on the test data results 

- The larger variety of data points your data set contains, the more complex model we can build without overfitting

- Score method returns R^2 value which is also known as coefficient of determination and is a measure of goodness of a prediction for a regression model. So basically whenever we have to see how well the predictions are for a data set we can use this function to check

- Parameters of KNN:
    - Numbers of neighbors
    - how we measure distance between the neighbors ( generally euclidean distance is used which works well in many settings )
    
- Strengths of KNN : 
    - Easy to understand 
    - often gives reasonable performance without a lot of adjustments
    - fast to implement 
    
- Weakness of KNN :
    - Very slow on large dataset to predict the values 
    - does not perform well on datasets with many features 

## Classification ( when output is a list of possible values)

- if the number of classes are two then it's a binary classification else multiclass classification


## Regression

- When we have to predict a continous number or a floating point number we use this model

### Definitions

- Overfit - when we make the model too complex, we start to focus on indivudual data points of training set, then it will be a overfit model

- Underfit - when we make the model too simple that even training set do badly on that, then it's underfit 

- Generalization - when the model can predict the results on unseen data then we say it is able to generalize from training set to the test set

- Sweet spot - being in the middle of overfit and underfit is the sweet spot to get best Generalization performance

```python
X,y = mglearn.datasets.make_forge()

# Creating the testing and training sets 
# HERE X_train are the values from which we will get the y_train which is basically the predicted output 
# we are feeding the system information so that we can use it for the later stages

X_train,X_test, y_train, y_test = train_test_split(X,y,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)
```

```python
clf.fit(X_train,y_train)
```

```python
# Predicting the output
print('Test Set Prediction : {}'.format(clf.predict(X_test)))
```

```python
# Accuracy 
print('Accuracy : {}'.format(clf.score(X_test,y_test)))
```

```python
cancer_df = load_breast_cancer()
X_train,X_test, y_train, y_test = train_test_split(cancer_df.data,cancer_df.target,random_state=66)

training_accuracy = []
test_accuracy = []
n = range(1,11)

for i in n:
    clf = KNeighborsClassifier(n_neighbors=i)
    clf.fit(X_train,y_train)
    training_accuracy.append(clf.score(X_train,y_train))
    test_accuracy.append(clf.score(X_test,y_test))
    
plt.plot(n, training_accuracy, label="training accuracy")
plt.plot(n, test_accuracy, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
```
